## 4.0 Visualisation

### 4.1 Revenue per customer segment

```python
# Set style
sns.set_style("whitegrid")

# Colour
sns.set_palette("deep")

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=revenue_per_segment, x='CustomerSegment', y='TotalRevenue')

# Set y axis format
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'£{int(x):,}'))

plt.title('Revenue per customer segment', fontsize=14, fontweight='bold')
plt.xlabel('Customer Segment')
plt.ylabel('Total revenue')
plt.xticks(rotation=45)
plt.show()
```
