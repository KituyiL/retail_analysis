## 2.0 Paths

```python
# Current directory
CURRENT_DIR = Path.cwd()

# Path to main folder
PROJECT_ROOT = CURRENT_DIR.parent.parent

# Path to folders
ZERO_PRICE_PATH = PROJECT_ROOT/"data"/"processed"/"zero_price.csv"
```

## 3.0 Load dataset

```python
zero_price = pd.read_csv(ZERO_PRICE_PATH)
zero_price
```

```python
# Shape of dataset
shape = zero_price.shape
print(f'Number of rows: {shape[0]}')
print(f'Number of columns: {shape[1]}')
```

## 4.0 Distinct products that had zero unit price

```python
# Number of distinct products
total_products = zero_price['Description'].nunique()

print(f'Total distinct products : {total_products}')
```

```python
# Number of times product has had zero unit price
distinct_products = zero_price.groupby('Description').agg(
    Count = ('Description', 'count'),
    TotalRevenue = ('TotalPrice', 'sum')
).reset_index().sort_values('Count', ascending=False).reset_index(drop=True)

#pd.set_option('display.max_rows', None) # To see all rows

print(distinct_products)
```
