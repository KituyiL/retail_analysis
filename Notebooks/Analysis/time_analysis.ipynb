# SQL RESULTS - Time

## 1.0 Import necessary libraries

```python
# Data analysis
import pandas as pd

# Directory
from pathlib import Path

# Visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2.0 Paths setup

```python
# Current directory
CURRENT_DIR = Path.cwd()

# Path to main folder
PROJECT_ROOT = CURRENT_DIR.parent.parent

# Path to folders and files
RESULTS_FOLDER = PROJECT_ROOT/"SQL_results"
TIME_ANALYSIS_RESULTS = RESULTS_FOLDER/"time_analysis"
```

## 3.0 Daily Analysis

### 3.1 Average daily revenue

```python
average_daily_revenue = pd.read_csv(TIME_ANALYSIS_RESULTS/"average_daily_revenue.csv")
average_daily_revenue
```

### 3.2 Average daily orders

```python
 average_daily_orders = pd.read_csv(TIME_ANALYSIS_RESULTS/"average_daily_orders.csv")
 average_daily_orders
```

### 3.3 Top average hourly revenue

```python
average_hourly_revenue = pd.read_csv(TIME_ANALYSIS_RESULTS/"average_hourly_revenue.csv")
average_hourly_revenue
```

### 3.4 Top average hourly orders

```python
average_hourly_orders = pd.read_csv(TIME_ANALYSIS_RESULTS/"average_hourly_orders.csv")
average_hourly_orders
```

## 4.0 Visualisation

### 4.1 Average daily revenue

```python
# Set style
sns.set_style("whitegrid")

# Colour
sns.set_palette("deep")

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=average_daily_revenue, x='Weekday', y='AvgDailyRevenue')
plt.title('Average Daily revenue', fontsize=14, fontweight='bold')
plt.xlabel('Day')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

### 4.2 Top average hourly revenue

```python
# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=average_hourly_revenue, x='Weekday', y='AvgHourlyRevenue', marker='s', label='Avg Hourly Revenue')
sns.lineplot(data=average_hourly_revenue, x='Weekday', y='PeakRevenue', marker ='o', label='Peak Revenue')
plt.title('Top average hourly revenue vs Peak revenue', fontsize=14, fontweight='bold')
plt.xlabel('Day')
plt.ylabel('Revenue')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

### 4.3 Top average hourly orders

```python
# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=average_hourly_orders, x='Weekday', y='AvgHourlyOrders', marker='s', label='Avg Hourly Orders')
sns.lineplot(data=average_hourly_orders, x='Weekday', y='PeakOrders', marker ='o', label='Peak Orders')
plt.title('Top average hourly orders vs Peak orders', fontsize=14, fontweight='bold')
plt.xlabel('Day')
plt.ylabel('Orders')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
