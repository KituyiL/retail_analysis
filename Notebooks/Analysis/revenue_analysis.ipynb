## 1.0 Import necessary libraries

```python
# Data analysis
import pandas as pd

# Directory
from pathlib import Path

# Visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## 2.0 Paths setup
```

```python
# Current directory
CURRENT_DIR = Path.cwd()

# Path to main folder
PROJECT_ROOT = CURRENT_DIR.parent.parent

# Path to folders
RESULTS_FOLDER = PROJECT_ROOT/"SQL_results"
REVENUE_ANALYSIS_RESULTS = RESULTS_FOLDER/"revenue_analysis"
```

## 3.0 Revenue Analysis

### 3.1 Total revenue

```python
total_revenue = pd.read_csv(REVENUE_ANALYSIS_RESULTS/"total_revenue.csv")
total_revenue
```

### 3.2 Monthly revenue

```python
monthly_revenue = pd.read_csv(REVENUE_ANALYSIS_RESULTS/"monthly_revenue.csv")

# Change Months to actual names
monthly_revenue['OrderMonth'] = pd.to_datetime(monthly_revenue['OrderMonth']).dt.strftime('%b %Y')

monthly_revenue
```

### 3.3 Top 5 months by revenue

```python
top_5_months_revenue = pd.read_csv(REVENUE_ANALYSIS_RESULTS/"top_5_months_revenue.csv")
top_5_months_revenue
```

### 3.4 Bottom 5 months by revenue

```python
bottom_5_months_revenue = pd.read_csv(REVENUE_ANALYSIS_RESULTS/"bottom_5_months_revenue.csv")
bottom_5_months_revenue
```

### 3.5 Revenue growth rate

```python
revenue_growth_rate = pd.read_csv(REVENUE_ANALYSIS_RESULTS/"revenue_growth_rate.csv")
revenue_growth_rate
```

## 4.0 Visualisation

### 4.1 Revenue per month

```python
# Set style
sns.set_style("whitegrid")

# Colour
sns.set_palette("deep")

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_revenue, x='OrderMonth', y='TotalRevenue')
plt.title('Revenue per month', fontsize=14, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```
