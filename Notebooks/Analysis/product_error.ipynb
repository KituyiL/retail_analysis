# Products with errors

## 1.0 Import nececssary libraries

```python
# Data analysis
import pandas as pd

# Directory
from pathlib import Path
```

## 2.0 Paths

```python
# Current directory
CURRENT_DIR = Path.cwd()

# Path to main folder
PROJECT_ROOT = CURRENT_DIR.parent.parent

# Path to folders
PRODUCT_ERROR_PATH = PROJECT_ROOT/"data"/"processed"/"product_error.csv"
```

## 3.0 Load dataset

```python
product_error = pd.read_csv(PRODUCT_ERROR_PATH)
product_error
```

```python
# Shape of dataset
shape = product_error.shape
print(f'Number of rows: {shape[0]}')
print(f'Number of columns: {shape[1]}')
```

## 4.0 Distinct products that had errors

```python
# Number of distinct errors
total_errors = product_error['Description'].nunique()

print(f'Total distinct errors : {total_errors}')
```

```python
# Number of times error has occurred
distinct_errors = product_error.groupby('Description').agg(
    Count = ('Description', 'count')
).reset_index().sort_values('Count', ascending=False).reset_index(drop=True)

#pd.set_option('display.max_rows', None) # To see all rows

print(distinct_errors)
```
