### 3.6 Top 3 products per month by sales

```python
top_3_per_month_by_sales = pd.read_csv(PRODUCT_ANALYSIS_RESULTS/"top_3_per_month_by_sales.csv")
top_3_per_month_by_sales
```

#### 3.6.1 Count of distinct products in 'top_3_per_month_by_sales'

```python
top_product_monthly_sales = top_3_per_month_by_sales.groupby('ProductDescription').agg(
    No_of_times = ('ProductDescription', 'count')
).reset_index().sort_values('No_of_times', ascending=False)
print(top_product_monthly_sales)
```

### 3.10 Top 3 products per country by sales

```python
top_3_per_country_by_sales = pd.read_csv(PRODUCT_ANALYSIS_RESULTS/"top_3_per_country_by_sales.csv")
top_3_per_country_by_sales
```

#### 3.10.1  Count of distinct products in 'top_3_per_country_by_sales'

```python
import textwrap
top_product_country_sales = top_3_per_country_by_sales.groupby('ProductDescription').agg(
    No_of_times = ('ProductDescription', 'count'),
    Countries=('Country', lambda x: '\n'.join(textwrap.wrap(', '.join(sorted(x.unique())), width=50)))
).reset_index().sort_values('No_of_times', ascending=False)

top_product_country_sales.style.set_properties(**{'text-align': 'left'})
```
