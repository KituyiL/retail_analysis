```python
# Data analysis
import pandas as pd

# Directory
from pathlib import Path
```

## 2.0 Paths setup

```python
# Current directory
CURRENT_DIR = Path.cwd()

# Path to main folder
PROJECT_ROOT = CURRENT_DIR.parent.parent

# Path to folders
CANCELLED_ORDERS_PATH = PROJECT_ROOT/"data"/"processed"/"cancelled_orders.csv"
```

## 3.0 Load dataset

```python
cancelled_orders = pd.read_csv(CANCELLED_ORDERS_PATH)
cancelled_orders
```

```python
# Shape of dataset
shape = cancelled_orders.shape
print(f'Number of rows: {shape[0]}')
print(f'Number of columns: {shape[1]}')
```

## 4.0 Distinct products that were cancelled

```python
# Number of distinct products
total_products = cancelled_orders['Description'].nunique()

print(f'Total distinct products : {total_products}')
```

```python
# Number of times the product was cancelled
distinct_products = cancelled_orders.groupby('Description').agg(
    Count = ('Description', 'count'),
    TotalRevenue = ('TotalPrice', 'sum')
).reset_index().sort_values('Count', ascending=False).reset_index(drop=True)

#pd.set_option('display.max_rows', None) # To see all rows

print(distinct_products)
```

```python
lost_revenue = cancelled_orders['TotalPrice'].sum()
lost_revenue
```
