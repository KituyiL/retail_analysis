### 3.4.1 Missing Customer Id

```python
# Confirm CustomerId is 5-digits
# Create dataframe copy 
cleaned_df2 = cleaned_df.copy()

# Customer ids that are not 5 digits
cleaned_df2['CustomerID_str'] = cleaned_df2['CustomerID'].astype(str)
odd_customerid = cleaned_df2[~cleaned_df2['CustomerID_str'].astype(str).str.fullmatch(r'\d{5}')]
odd_customerid
```

```python
# Confirm the odd customer ids are the same as null customer ids
# Missing customer id
null_customerid = cleaned_df[cleaned_df['CustomerID'].isnull()]

# Confirm
odd_customerid.drop(columns='CustomerID_str').equals(null_customerid)
```

```python
# Fill null customer id if invoice number has at least 1 customer id
# Groups rows by invoice No and fill missing customerIds forwads and backwards
cleaned_df = cleaned_df.copy()
cleaned_df["CustomerID"] = cleaned_df.groupby("InvoiceNo")["CustomerID"].transform(lambda x: x.ffill().bfill())

# Check for missing values
cleaned_df.isnull().sum()
```

```python
# Turn  Stock codes with no description to dataframe
nodescription_df = pd.DataFrame({'StockCode': no_description})
nodescription_df
```

```python
# Function to create pattern for stock code. D for numbers and L for letters
def get_pattern(code_numbers):
    code_numbers = str(code_numbers).strip().upper()
    pattern = ''
    for char in code_numbers:
        if char.isdigit():
            pattern += 'D'
        elif char.isalpha():
            pattern += 'L'
        else:
            pattern += '?'
    return pattern

nodescription_df['Pattern'] = nodescription_df['StockCode'].apply(get_pattern)
```

```python
# categorize the patterns
import re
def group_pattern(p):
    if set(p) == {'D'}:
        return 'Numeric'
    elif re.fullmatch(r'D+L+', p):
        return 'Letter at end'
    elif re.fullmatch(r'L+D+', p):
        return 'Letter at start'
    elif re.fullmatch(r'[DL]+', p):
        return 'Mixed'
    else:
        return 'Other'
        
    
    
nodescription_df['Category'] = nodescription_df['Pattern'].apply(group_pattern)

nodescription_df[['StockCode', 'Pattern', 'Category']].head(20)
```

```python
# Number of stock codes per pattern group
nodescription_df['Pattern'].value_counts()
```

```python
# exclude digit only stockcode from nodescription_df
nodescription_df2 = nodescription_df[nodescription_df['Pattern'] != 'DDDDD'].copy()

# remove letters to remain with numbers
nodescription_df2['Numeric'] = nodescription_df2['StockCode'].astype(str).str.replace(r'[A-za-z]', '', regex=True)

# set of all stockcodes in cleaned df
allcodes = set(original_df['StockCode'].astype(str))

# filter odd stock codes that have a numeric counterpart
nodescription_df2['NumericMatches'] = nodescription_df2['Numeric'].isin(allcodes)

# codes that have a match
matched_df = nodescription_df2[nodescription_df2['NumericMatches']][['StockCode', 'Numeric']].astype(str).drop_duplicates()

matched_df.head()
```

```python
# Drop the remaining null values
cleaned_df = cleaned_df.dropna()
cleaned_df.isnull().sum()
```
